```python
delay_data = x[['DAY_OF_MONTH', 'DAY_OF_WEEK', 'OP_CARRIER_AIRLINE_ID', 'OP_CARRIER_FL_NUM', 'ORIGIN_AIRPORT_ID',
                'ORIGIN_AIRPORT_SEQ_ID', 'ORIGIN_CITY_MARKET_ID', 'DEST_AIRPORT_ID', 'DEST_AIRPORT_SEQ_ID', 
                'DEST_CITY_MARKET_ID', 'CRS_DEP_TIME', 'CRS_ARR_TIME', 'DISTANCE', 'DEP_TIME', 'DEP_DELAY',
                'DEP_DELAY_NEW', 'ARR_TIME', 'ARR_DELAY', 'ARR_DELAY_NEW', 'CARRIER_DELAY', 'WEATHER_DELAY', 
                'NAS_DELAY', 'SECURITY_DELAY', 'LATE_AIRCRAFT_DELAY']]
# 'OP_UNIQUE_CARRIER', 'TAIL_NUM'
```

```python
delay_data.dropna(inplace=True)
delay_data.describe()
```

```python
delay_data = delay_data.astype({'OP_CARRIER_AIRLINE_ID': int, 'ORIGIN_AIRPORT_SEQ_ID': int, 'ORIGIN_AIRPORT_ID': int,
              'ORIGIN_CITY_MARKET_ID': int, 'DEST_AIRPORT_SEQ_ID': int, 'DEST_AIRPORT_ID': int,
              'DEST_CITY_MARKET_ID': int})
delay_data.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
```

```python
X = delay_data.iloc[:, :13]
Y = delay_data.iloc[:, 13]
print(X,'\n',Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.10, random_state=2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
print(X_train)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
print(f"intercept: {lr_model.intercept_}")
print(f"coefficients: {lr_model.coef_}")
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
dtr_model = DecisionTreeRegressor()
dtr_model.fit(X_train, y_train)
y_pred = dtr_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
print(f"R-squared score: {r2}")
print(np.sqrt(mean_squared_error(y_test,y_pred)))
dt=mean_squared_error(y_test,y_pred )
print(mean_squared_error(y_test,y_pred ))
print(mean_absolute_error(y_test,y_pred ))
```
